```python
import json
import pandas as pd
from IPython.display import display
pd.set_option('display.precision', 10)
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
# %matplotlib qt

uuid = "944A16A5-9D5D-4348-B532-10B926BB86A1" # up / down wrist motion
uuid = "2662B396-20B2-4EF4-903D-915CDD2660C0" # Around and then over a bit
# uuid = "8AA26238-1BAA-4C00-82E2-6E5EE496A0DE"
# uuid = "A4E3E98D-2110-4C75-8C4D-AADC0E5F6B2D"
uuid = "8E101B8C-D588-4416-BA98-678D218CA125"
uuid = "792DBF78-CCE2-44A2-BC1D-D618789F099A"
uuid = "99AAC7DE-4106-43AB-AA80-F5816AB4DB85"

def import_uuid(uuid):
    
    # Define the file path
    raw_data_filepath = "Data/" + uuid + ".json"
    timestamps_filepath = "Data/Timestamps_" + uuid + ".json"
    
    raw_df = import_data_json(raw_data_filepath)
    timestamps_df = import_timestamp_json(timestamps_filepath)
    return (raw_df, timestamps_df)

    

def import_data_json(filepath):
    # Read the JSON data from the file
    with open(filepath, 'r') as file:
        data = json.load(file)

    # Extract workoutDatas
    workout_datas = data["workoutDatas"]

    # Create a DataFrame
    df = pd.json_normalize(workout_datas)
    return df

def import_timestamp_json(timestamps_filepath):
    # return pd.read_json(timestamps_filepath)
    with open(timestamps_filepath, 'r') as file:
        data = json.load(file)

    # Step 2: Structure the data
    # Create a dictionary to hold structured data
    structured_data = {
        "startTime": data["startTime"],
        "endTime": data["endTime"],
        "groundLeaveTimes": data["groundLeaveTimes"],
        "airLeaveTimes": data["airLeaveTimes"]
    }

    # Step 3: Create the DataFrame
    # If you want to have each time in a separate row, you can normalize the lists
    df_ground = pd.DataFrame(structured_data["groundLeaveTimes"], columns=["groundLeaveTimes"])
    df_air = pd.DataFrame(structured_data["airLeaveTimes"], columns=["airLeaveTimes"])

    # Combine into a single DataFrame if needed
    df = pd.concat([df_ground, df_air], axis=1)

    # Add startTime and endTime to every row if needed
    df["startTime"] = structured_data["startTime"]
    df["endTime"] = structured_data["endTime"]
    return df
    
raw_df, timestamps_df = import_uuid(uuid)
mask = (raw_df['time'] >= timestamps_df['startTime'][0]) * (raw_df['time'] <= timestamps_df['endTime'][0])
trimmed_raw_df = raw_df[mask]
trimmed_raw_df["time"] = trimmed_raw_df["time"].apply(lambda x: x - timestamps_df["startTime"][0])
timestamps_df["airLeaveTimes"] = timestamps_df["airLeaveTimes"].apply(lambda x: x - timestamps_df["startTime"][0])
end_time = timestamps_df["endTime"][0] - timestamps_df["startTime"][0]
display(trimmed_raw_df)
display(timestamps_df["airLeaveTimes"])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error

# Load the dataset
def load_data(file_path):
    data = pd.read_csv(file_path)
    return data

# Preprocess the data
def preprocess_data(datasets, timestamp_datasets):
    evaluated_datasets = []
    evaluated_timestamps = []
    for dataset in datasets:
        X = dataset.values.reshape(-1,10)
        evaluated_datasets.append(X)
    # for timestamps in timestamp_datasets:
    #     y = timestamps["airLeaveTimes"].values
    #     evaluated_timestamps.append(y)
    # print(y)
    return evaluated_datasets, timestamp_datasets

# Extract features (in this case, we will use position as the feature directly)
def extract_features(X):
    return X

# Train the model
def train_model(X_train, y_train):
    scaler_X = StandardScaler()
    scaler_y = StandardScaler()

    X_train_scaled = scaler_X.fit_transform(X_train)
    y_train_scaled = scaler_y.fit_transform(y_train.reshape(-1, 1)).ravel()

    model = SVR(kernel='rbf', C=1e3, gamma=0.1)
    model.fit(X_train_scaled, y_train_scaled)

    return model, scaler_X, scaler_y

# Predict timestamps
def predict_timestamps(model, scaler_X, scaler_y, X_test):
    X_test_scaled = scaler_X.transform(X_test)
    y_pred_scaled = model.predict(X_test_scaled)
    y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 1))
    return y_pred
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, LSTM, Dense, TimeDistributed, RepeatVector
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split

print("Num GPUs Available: ", len(tf.config.experimental.list_physical_devices('GPU')))
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        logical_gpus = tf.config.experimental.list_logical_devices('GPU')
        print(len(gpus), "Physical GPUs,", len(logical_gpus), "Logical GPUs")
    except RuntimeError as e:
        print(e)

# Assuming X is your input data (1000 samples of 10x600 matrices)
# and Y is your output data (list of variable-length sequences)
# X = np.random.rand(1000, 10, 600, 1)  # Example input data
# Y = [np.random.rand(np.random.randint(1, 50), 1) for _ in range(1000)]  # Example variable-length output sequences
X, Y = preprocess_data(datasets_7s_training, timestamps_7s_training)
print(np.shape(X))
# X = np.random.rand(1000, 10, 600, 1)  # 1000 samples of 10x600 matrices with 1 channel
# Y = [np.random.rand(np.random.randint(1, 50), 1) for _ in range(1000)]  # Example variable-length output sequences

# Padding the output sequences
max_seq_length = max(len(seq) for seq in Y)
Y_padded = pad_sequences(Y, maxlen=max_seq_length, padding='post', dtype='float32')
num_samples = len(X)
Y_padded = Y_padded.reshape(num_samples, max_seq_length, 1)
X = np.array(X).reshape(num_samples, 10, 700, 1)
print(np.shape(Y_padded))
# print(Y_padded)

# Splitting data into training and validation sets
X_train, X_val, Y_train, Y_val = train_test_split(X, Y_padded, test_size=0.2, random_state=42)

# Verify that the split is done correctly
print(f"X_train shape: {np.shape(X_train)}")
print(f"Y_train shape: {np.shape(Y_train)}")
print(f"X_val shape: {np.shape(X_val)}")
print(f"Y_val shape: {np.shape(Y_val)}")

# Define the model
model = Sequential([
    Conv2D(128, (3, 3), activation='relu', input_shape=(10, 700, 1)),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    RepeatVector(max_seq_length),
    LSTM(64, return_sequences=True),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')

# Training the model
print("Training model")
model.fit(X_train, Y_train, validation_data=(X_val, Y_val), epochs=300, batch_size=1000)

# Evaluate the model (if you have a test set)
# X_test, Y_test = ...  # Your test data
loss = model.evaluate(X_val, Y_val)
print(f'Test Loss: {loss}')

```
